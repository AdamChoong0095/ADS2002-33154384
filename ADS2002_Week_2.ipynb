```python
import pandas as pd
import numpy as np
cars=pd.read_csv('/content/drive/MyDrive/mtcars.csv')
cars.head()
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
X=cars[['wt','hp']]
y=cars['mpg']
linear=LinearRegression(fit_intercept=True)
linear.fit(X,y)
preds=linear.predict(X)
print("Coefficients are",np.round(linear.coef_, 3))
print("Intercept is",np.round(linear.intercept_,3))
print("R2 score is ", np.round(r2_score(y,preds),3))
```
